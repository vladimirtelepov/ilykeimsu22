```python
from autogluon.tabular import TabularDataset, TabularPredictor
train_data = TabularDataset('./train_sample.csv')
test_data = TabularDataset('./test_sample.csv')
```

```python
label = 'target'
eval_metric = 'log_loss'
time_limit = 1800
```

```python
predictor = TabularPredictor(
    label=label, eval_metric=eval_metric, learner_kwargs={'ignored_columns': ['id']}
).fit(train_data, presets='best_quality', time_limit=time_limit)
```

```python
predictor.leaderboard(silent=True)
```

```python
import pandas as pd
submission = test_data[['id']]
test_pred_proba = predictor.predict_proba(test_data)
submission = pd.concat([submission, test_pred_proba], axis=1)
submission.to_csv('result.csv', index=False)
submission.head()
```
